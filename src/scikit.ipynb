```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error
```

```python
houses= pd.read_csv('../datasets/Housing.csv')
houses.head()
```

```python
X_train= houses.loc[:, 'area': 'stories'].to_numpy()
Y_train= houses['price'].to_numpy().reshape(-1, 1)
```

```python
scaler= StandardScaler()
X_norm= scaler.fit_transform(X_train)
Y_norm= scaler.fit_transform(Y_train)
print(f'Range in original data: {np.ptp(X_train, axis= 0)}')
print(f'Range in normalized data: {np.ptp(X_norm, axis=0)}')
```

```python
model= LinearRegression()
model.fit(X_norm, Y_norm)
W= model.coef_
b= model.intercept_
print(f'W={W}\nb={b}')
```

```python
y_pred_norm= model.predict(X_norm)
y_pred_norm[0:5]
```

```python
y_norm_calc= X_norm@ W.reshape(-1,1) + b
y_norm_calc[0:5]
```
